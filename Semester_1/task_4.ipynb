Задача 2.1. Брошены три монеты. Найти вероятности событий:
1. Первая монета выпала гербом.
2. Выпало ровно два герба.
3. Выпало не больше двух гербов.

```python
# Пункт 1
# Будем моделировать бросок монеты генерацией случайного числа -- 0 или 1. Пусть 0 обозначает выпадение решки, а 1 -- выпадение герба
n_experiments = 100000
success_count = 0

for i in range(n_experiments):
    mint_1 = random.randint(0, 1)
    mint_2 = random.randint(0, 1)
    mint_3 = random.randint(0, 1)
    if mint_1 == 1:
        success_count += 1

print(f'Оценка вероятности: {success_count / n_experiments}')
print('Истинная вероятность: 0.5')
```

Задача 2.7. Бросаются 4 игральные кости. Найти вероятность того, что на них выпадет по одинаковому числу очков.

```python
n_experiments = 100000
success_count = 0

for i in range(n_experiments):
    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)
    dice_3 = random.randint(1, 6)
    dice_4 = random.randint(1, 6)
    if dice_1 == dice_2 == dice_3 == dice_4:
        success_count += 1

print(f'Оценка вероятности: {success_count / n_experiments}')
print('Истинная вероятность: 0.00463')
```

Задача 3.7. На отрезок [0, a] наудачу брошены три точки. Найти вероятность того, что из отрезков, равных расстояниям от точки 0 до точки падения, можно составить треугольник.

```python
# Пункт 1

success_count = 0
n_experiments = 100000
a = 1

for i in range(n_experiments):
    x = random.uniform(0, a)
    y = random.uniform(0, a)
    z = random.uniform(0, a)
    if (x + y >= z) and (x + z >= y) and (y + z >= x):
        success_count += 1

print(f'Оценка вероятности: {success_count / n_experiments}')
print('Истинная вероятность: 0.5')
```
