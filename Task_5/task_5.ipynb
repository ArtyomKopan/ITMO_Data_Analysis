```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('beverage_r.csv', sep=';')
```

Задача 1

Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
train_df, test_df = train_test_split(df, train_size=0.66, test_size=0.33, random_state=42)
```

Задача 2

Разделить таблицу данных на четыре одинаковых части.

```python
df_1, df_2 = train_test_split(df, test_size=0.5, random_state=42)
df_1, df_3 = train_test_split(df_1, test_size=0.5, random_state=42)
df_2, df_4 = train_test_split(df_2, test_size=0.5, random_state=42)
```

Задача 3

Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
df_1, df_2 = train_test_split(df, test_size=0.5, stratify=df['COKE'], random_state=42)
df_3, df_4 = train_test_split(df, test_size=0.5, stratify=df['TAB'], random_state=42)
```
